## ZHVI Single_Family Homes Time Series ($)
https://www.zillow.com/research/data/


```python
# import libarary

import pandas as pd
```

```python
# get and read datasource

df = pd.read_csv('https://files.zillowstatic.com/research/public_csvs/zhvi/State_zhvi_uc_sfr_tier_0.33_0.67_sm_sa_month.csv?t=1675567268', low_memory=False)
df.head()
```

```python
# check detail of the data

df.info()
```

```python
# find how many Region in the Dataset

df['RegionName'].value_counts().sum()
```

```python
# showing the name of each Region

df['RegionName'].unique()
```

```python
# there is 'NaN' value in the first 5th dataset, so let's check all the value or other value in the 'StateName' Columns 

df['StateName'].unique()
```

```python
# beacuse all the value is 'NaN', so we don't need it anymore and we can delete that columns
# and let's check the difference

df = df.drop(columns='StateName')
df.head()
```

```python
# because we have 5 columns, and 280 columns (too many columns), it's look messy and not a good datastructure
# so now, we will rearrange this dataset into new position using .melt() pandas method  

df = pd.melt(frame=df, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType'], var_name='year-month', value_name='count')
```

```python
# let's check the difference 

df.head(2)
```

```python
# now we can see that now only 6 columns and 14076 rows, and much easier to understand or process the dataset

df.shape
```

```python
# now we'll trying to separate day, month and year into its own columns
# separate this using str.split and then add dash'-' as indicator

df['year-month'].str.split('-').head()
```

```python
# place the new 3 columns into dataset

df[['year', 'month', 'day']] = df['year-month'].str.split('-', expand=True)
```

```python
df.head()
```

```python
# we can see that we don't need 'year-month' columns anymore
# so now we can delete it, but before doing that we can copy the original dataset for backup if there is something wrong happen

df1 = df.copy(deep=True)
df1.head(2)
```

```python
# next, use .drop() method to delete 'year-month' columns
# and done. we got it

df1 = df1.drop(columns=['year-month'])
df1.head(2)
```
